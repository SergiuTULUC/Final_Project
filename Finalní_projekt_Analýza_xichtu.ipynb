```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import cv2
from deepface import DeepFace
```

### Načtení obrázku s vělkémá, hluboce otevřenéma očíma

```python
img = cv2.imread('0029.jpg') # načítam obrazek
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # konverze barev z BGR na RGB
img
```

zobrazení obrázku na plotu


```python
plt.imshow(img)
plt.show()
```

Analýza obrázku

```python
prediction = DeepFace.analyze(img)
```

```python
#Ukaž mi vysledky analýzy
prediction
```

```python
#Co má za naladu
prediction[0]['dominant_emotion'] # Access the first element (dictionary) of the list and then access the 'dominant_emotion' key.
```

# Vyhledavaní obličeje na obrázku a označení čvercem

```python
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)


```

```python
#Zobrazení obrázku v černobílém formátu
plt.imshow(gray, cmap='gray')
#plt.show()
```

### Vyhledavame obličej a oramujeme

```python
face = face_cascade.detectMultiScale(gray, 1.3, 6)
# Kreslime rameček
for(x, y, w, h) in face:
    cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 0), 2)
```

```python
# Výsledek

plt.imshow(img)
#plt.show()
```
